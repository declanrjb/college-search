```python
import pandas as pd
import os
import glob
import re
import zipfile
import shutil
```

```python
os.mkdir('../data/ipeds-clean')
os.mkdir('../data/ipeds-clean/admissions')
os.mkdir('../data/ipeds-clean/enrollment')
```

```python
files = list(glob.iglob('../data/ipeds/**/**.zip', recursive=True))
```

```python
for file in files:
    match_obj = re.search('/ADM[0-9]{4}.zip', file)
    if match_obj:
        stem = file.split('/')[-1]
        shutil.copy(file, f'../data/ipeds-clean/admissions/{stem}')
```

```python
for file in files:
    match_obj = re.search('/EFFY[0-9]{4}.zip', file)
    if match_obj:
        stem = file.split('/')[-1]
        shutil.copy(file, f'../data/ipeds-clean/enrollment/{stem}')
```

unzip clean files

```python
files = list(glob.iglob('../data/ipeds-clean/**/**.zip', recursive=True))
```

```python
for file in files:
    out_dir = '/'.join(file.split('/')[:-1])
    with zipfile.ZipFile(file, 'r') as zip_ref:
        zip_ref.extractall(out_dir)
    os.remove(file)
```
