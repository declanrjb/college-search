```python
import pandas as pd
import os
import glob
import re
```

```python
files = list(glob.iglob('../data/ipeds/**/**.zip', recursive=True))
```

```python
for file in files:
    match_obj = re.search('/ADM[0-9]{4}.zip', file)
    if match_obj:
        stem = file.split('/')[-1]
        os.rename(file, f'../data/ipeds-clean/admissions/{stem}')
```

```python
for file in files:
    match_obj = re.search('/EFFY[0-9]{4}.zip', file)
    if match_obj:
        print(file)
        stem = file.split('/')[-1]
        os.rename(file, f'../data/ipeds-clean/enrollment/{stem}')
```
